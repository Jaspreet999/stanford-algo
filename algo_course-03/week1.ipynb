# prims algorithm

```python
class Adj:
    
    def __init__(self,d,dis):
        self.d = d
        self.dis = dis

class Graph:
    
    def __init__(self,nodes):
        self.nodes = nodes
        self.adj = {}
        
        for i in range(1,self.nodes+1):
            self.adj[i] = []
    
    def addEdge(self,s,d,dis):
        self.adj[s].append(Adj(d,dis))
        self.adj[d].append(Adj(s,dis))
    
    def print_graph(self):
        for i in self.adj:
            print(i,end=" -> ")
            for adj in self.adj[i]:
                print(adj.d,adj.dis,end=", ")
            print('\n') 
            

# graph = Graph(4)
# graph.addEdge(1,2,1)
# graph.addEdge(2,4,2)
# graph.addEdge(3,1,4)
# graph.addEdge(4,3,5)
# graph.addEdge(4,1,3)
# graph.print_graph()
```

```python
op1 = open('prims_edges.txt','r')

temp = []

for i in op1:
    temp.append(i.split(' '))
    
graph = Graph(int(temp[0][0]))
    
temp.pop(0)

for i in temp:
    graph.addEdge(int(i[0]),int(i[1]),int(i[2]))
```

```python
N = graph.nodes

visited = []

heap = [200000000]*(N+1)

heap[1] = 0
    
minCost = 0

while len(visited) != N:
    
    dis = min(heap)
    node = heap.index(dis)
    visited.append(node)
    minCost += dis
    
    for i in graph.adj[node]:
        if i.d not in visited:
            heap[i.d] = min(heap[i.d] , i.dis)
    
    heap[node] = 200000000000
print(minCost)    
```
