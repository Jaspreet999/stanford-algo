```python
import random 

def kargar(graph,nodes):
     
    cut = len(nodes)
    
    while(cut>2):
        random.shuffle(nodes)
        
        s = random.choice(nodes)
        d = random.choice(nodes)
        
        while(d not in graph.adj_list[s]) :
            s = random.choice(nodes)
            d = random.choice(nodes)
            

        #get the adjacent nodes
        c = graph.adj_list[s]
        e = graph.adj_list[d]
#         print(c,e,s,d)
        #remove cycle
        c = c+e
        
        
        i = len(c) -1
        while(i>=0):
            if(s == c[i] or d == c[i]):
                c.remove(c[i])
            i-=1
            
            
        graph.adj_list[s] = c

        del graph.adj_list[d]

        #change d with s in all the nodes
        for node in graph.adj_list:
            adj = graph.adj_list[node]
            for i in adj:
                if d == i:
                    adj[adj.index(i)] = s
        
        nodes = list(graph.adj_list)
        
        cut -= 1

        
    return len(graph.adj_list[nodes[0]])
          
    
#graph creation    
class Graph:
    def __init__(self,Nodes):
        self.nodes = Nodes
        self.adj_list = {}
        
        for node in self.nodes:
            self.adj_list[node] = []
            
    def add_edge(self,s,d):
        self.adj_list[s].append(d)
     
    def print_graph(self):
        for node in self.adj_list:
            print(node,"->",self.adj_list[node])
```
