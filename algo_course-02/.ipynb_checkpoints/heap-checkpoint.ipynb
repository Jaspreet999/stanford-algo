```python
class MinHeap:
    
    def __init__(self,size):
        self.size = size
        self.point = 0
        self.arr = [-1]*size
        
    def parent(self,n):
        return int((n-1)/2)
        
        
    def leftChild(self,n):
        return int(2*n + 1)
        
    def rightChild(self,n):
        return int(2*n + 2)
        
        
    def insert(self,key):
        
        if(self.point == self.size) : return -1
        
        self.arr[self.point] = key
        
        self.maintainInsert()
        
        self.point += 1
    
    def maintainInsert(self):
        if(self.point == 0): return 
        
        i = self.point
        
        while(i > 0 and self.arr[self.parent(i)] > self.arr[i]):
            temp = self.arr[self.parent(i)]
            self.arr[self.parent(i)] = self.arr[i]
            self.arr[i] = temp
            
            i = self.parent(i)
      
    def minExtract(self):
        
        if(self.point <= 0):return -1
        
        if(self.point == 1):
            temp = self.arr[0]
            self.point -=1
            return temp
        else:
            i = self.point - 1
            temp = self.arr[0]
            self.arr[0] = self.arr[i]
            self.arr[i] = temp
            
            self.point -=1
            
            self.maintainDelete()
            
            return temp
        
    def maintainDelete(self):
        
        i = self.point
        if(i == 1): return
        j = 0
        l = self.leftChild(j)
        r = self.rightChild(j)
        
        while(j < self.point and l < self.point and r < self.point):
            top =l if self.arr[l]<self.arr[r] else r
            
            if(self.arr[j] > self.arr[top]):
                
                temp = self.arr[j]
                self.arr[j] = self.arr[top]
                self.arr[top] = temp
                
                j = top
                l = self.leftChild(top)
                r = self.rightChild(top)
            
            else:
                return
            
    def getMin(self):
        if self.point == 0: return -1
        return self.arr[0]
         
        
[5,9,12,13,16,45]       
heap = MinHeap(6) 
heap.insert(5)
heap.insert(6)
heap.insert(12)
heap.insert(13)
heap.insert(16)
heap.insert(45)

print(heap.arr)
print(heap.minExtract())
print(heap.minExtract())
heap.
print(heap.minExtract())
print(heap.minExtract())
print(heap.minExtract())
print(heap.minExtract())
print(heap.minExtract())
print(heap.minExtract())
print(heap.getMin())
```

```python
class MaxHeap:
    
    def __init__(self,size):
        self.size = size
        self.point = 0
        self.arr = [-1]*size
        
    def parent(self,n):
        return int((n-1)/2)
        
        
    def leftChild(self,n):
        return int(2*n + 1)
        
    def rightChild(self,n):
        return int(2*n + 2)
        
        
    def insert(self,key):
        
        if(self.point == self.size) : return -1
        
        self.arr[self.point] = key
        
        self.maintainInsert()
        self.point += 1
        
        return
    
    def maintainInsert(self):
        if(self.point == 0): return 
        
        i = self.point
        
        while(i > 0 and self.arr[self.parent(i)] < self.arr[i]):
            temp = self.arr[self.parent(i)]
            self.arr[self.parent(i)] = self.arr[i]
            self.arr[i] = temp
            
            i = self.parent(i)
        return    
      
    def maxExtract(self):
        
        if(self.point <= 0):return -1
        
        if(self.point == 1):
            temp = self.arr[0]
            self.point -=1
            return temp
        else:
            i = self.point - 1
            temp = self.arr[0]
            self.arr[0] = self.arr[i]
            self.arr[i] = -1
            
            self.maintainDelete()
            self.point -=1
            return temp
        
    def maintainDelete(self):
        
        i = self.point
        if(i == 1): return
        j = 0
        l = self.leftChild(j)
        r = self.rightChild(j)
#         print(l,r,i)
        while(j < self.point and l < self.point and r < self.point):
            top =l if self.arr[l]>self.arr[r] else r
            
            if(self.arr[j] < self.arr[top]):
                
                temp = self.arr[j]
                self.arr[j] = self.arr[top]
                self.arr[top] = temp
                
                j = top
                l = self.leftChild(top)
                r = self.rightChild(top)
            
            else:
                return
            
#         print(self.arr)    
            
    def getMax(self):
        if self.point == 0: return -1
        return self.arr[0]
         
        
heap = MaxHeap(7) 
heap.insert(100)
heap.insert(15)
heap.insert(30)
heap.insert(40)
heap.insert(50)
heap.insert(40)
heap.insert(10)
print(heap.arr)
print(heap.maxExtract())
print(heap.maxExtract())
print(heap.maxExtract())
print(heap.maxExtract())
print(heap.maxExtract())
print(heap.maxExtract())
print(heap.maxExtract())
print(heap.maxExtract())
print(heap.getMax())
```
