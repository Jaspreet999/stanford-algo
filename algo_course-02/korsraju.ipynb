```python
class Graph:
    
    def __init__(self,nodes):
        self.nodes = nodes
        self.adj_nodes = {}
        self.rev_adj_nodes = {}
        
        for i in nodes:
            self.adj_nodes[i] = []
            self.rev_adj_nodes[i] = []
    
    def add_edge(self,s,d):
        self.adj_nodes[s].append(d)
        
        #for reverse graph
        self.rev_adj_nodes[d].append(s)
        
    
    def print_graph(self):
        
        for i in self.adj_nodes:
            print(i,"->",end = " ")
            for j in self.adj_nodes[i]:
                print(j,end =" ")
            print()
            
    def reverse_graph(self):
        for i in self.rev_adj_nodes:
            print(i,"->",end = " ")
            for j in self.rev_adj_nodes[i]:
                print(j,end =" ")
            print()         
        
       
            
nodes=[1,2,3,4,5,6,7,8,9]        
graph = Graph(nodes)
graph.add_edge(1,4)
graph.add_edge(7,1)
graph.add_edge(4,7)
graph.add_edge(9,7)
graph.add_edge(6,9)
graph.add_edge(3,6)
graph.add_edge(9,3)
graph.add_edge(8,6)
graph.add_edge(8,5)
graph.add_edge(5,2)
graph.add_edge(2,8)


# graph.reverse_graph()

print(kosaraju(graph,nodes))
#graph.print_graph()
```

```python
def kosaraju(graph,nodes):
    
    node = nodes 
    #bcz first node start from 1
    visited = [False]*(len(node)+1)
    visited[0] = 0
    
    ft = {}
      
    for  i in range(len(node),0,-1):
        if visited[i] is False:
            dfs(i,graph,ft,visited)
            
    ft = sorted(ft.items(), key= lambda kv:(kv[1]), reverse=True)        

    fa = []
    
    for i in range(1,len(visited)):
        visited[i] =False
    
    
    for i in ft:
        if visited[i[0]] is False:
            fa.append(dfs2(i[0],graph,fa,visited))
    
    return fa

def dfs2(a,graph,fa,visited):
    
    stack = []
    stack.append(a)
    ans = 1
    
    while len(stack)!=0:
        i = stack[-1]
        flag = 0
        visited[i] = True
        
        for k in graph.adj_nodes[i]:
            if visited[k] is False:
                ans +=1
                flag = 1
                stack.append(k)
                break
                
        if flag == 0:
            stack.pop()
        
    return ans

def dfs(a,graph,ft,visited):
    
    stack = []
    stack.append(a)
    visited[a] = True
    
    while len(stack) != 0:
        a = stack[-1]
        visited[a] = True
        temp = 0
        
        for k in graph.rev_adj_nodes[a]:
            if visited[k] is False:
                temp = 1
                stack.append(k)

        if temp == 0:
            visited[0] +=1
            ft[stack[-1]] = visited[0]
            stack.pop()
        
```
